```python
# Реализуйте функцию train для нашего задания (по аналогии с частью 1)
# Функция должна строить графики loss и accuracy для train и test
# Не забудьте сделать ravel!

def mnist_train_verbose(network, train_dataloader, test_dataloader, epochs):
    optimizer = torch.optim.Adam(network.parameters(), lr=0.001)
    nll_loss = NLLLoss()
    ##### YOUR CODE HERE #####
    loss_hist_train = []
    loss_hist_test = []
    train_accuracy = []
    test_accuracy = []
    for i in range(epochs):
        # Training
        k = 0
        batch_size = 0
        for net in network:
            net.training = True
        for _, sample_batched in enumerate(train_dataloader):
            X_train = sample_batched[0]
            y_train = sample_batched[1].cuda()
            X_train = np.array(X_train).reshape(y_train.shape[0], -1)
            X_train = torch.from_numpy(X_train).cuda()
            y_pred = network(X_train)
            loss_train = nll_loss(y_pred, y_train)
            loss_hist_train.append(loss_train)
            
            
            batch_size = y_train.shape[0]
            num_correct = torch.sum(y_pred.argmax(1) == y_train)
            
            optimizer.zero_grad()
            loss_train.backward()
            optimizer.step()
        
        train_accuracy.append(float(num_correct) / batch_size)
            
        
        loss = 0
        for net in network:
            net.training = False
            
        for _, sample_batched in enumerate(test_dataloader):
            X_test = sample_batched[0]
            y_test = sample_batched[1].cuda()
            X_test = np.array(X_test).reshape(y_test.shape[0], -1)
            X_test = torch.from_numpy(X_test).cuda()
            y_pred_test = network(X_test)
            k+=1
            loss += nll_loss(y_pred_test, y_test)
            
            batch_size = y_test.shape[0]
            num_correct = torch.sum(y_pred_test.argmax(1) == y_test)
            
        test_accuracy.append(float(num_correct) / batch_size)    
        loss_hist_test.append(loss/k)

    fig = plt.figure()
    plt.title('train loss')
    plt.plot(range(len(loss_hist_train)), loss_hist_train)
    fig = plt.figure()
    plt.title('train accuracy')
    plt.plot(range(len(train_accuracy)), train_accuracy)
    fig = plt.figure()
    plt.title('test loss')
    plt.plot(range(len(loss_hist_test)), loss_hist_test)   
    fig = plt.figure()
    plt.title('test accuracy')
    plt.plot(range(len(test_accuracy)), test_accuracy)
```

Реализуйте DropConnect. 

```python
# полезная функция: .bernoulli_(p)
# не забывайте делать requires_grad=False у маски
# Не забудьте заменить все вероятности на p при test-проходе!

class DropConnect(nn.Module):
    def __init__(self, input_dim, output_dim, p=1.0, mean=0, std=0.01):
     ##### YOUR CODE HERE #####

        super(DropConnect, self).__init__()

        W = torch.FloatTensor(input_dim, output_dim).normal_(mean=mean, std=std)
        W.cuda()
        b = torch.zeros(output_dim)
        b.cuda()
        self.p = p
        self.b = torch.nn.Parameter(b)
        self.W = torch.nn.Parameter(W)
        

    def forward(self, x):
         ##### YOUR CODE HERE #####
        ones = torch.ones(self.W.shape, requires_grad=False).cuda()
#         ones = torch.ones(self.W.shape).cuda()
        mask = self.p*ones
        mask.cuda()
        if self.training:
            mask = torch.bernoulli(self.p * torch.ones(self.W.shape)).cuda()
            mask.requares_grad = False
        out = torch.mm(x, mask* self.W) + self.b
        return out
```

**Нормализация**

Давайте рассмотрим Batch Normalization. Этот слой вычитает среднее и делит на стандартное отклонение. Среднее и дисперсия вычисляются по батчу независимо для каждого нейрона. У этого слоя есть две важные проблемы: его нельзя использовать при обучении с размером батча 1 и он делает элементы батча зависимыми. Давайте реализуем аналог батч нормализации: <a href=https://arxiv.org/pdf/1607.06450.pdf>Layer normalization</a>. В layer normalization среднее и дисперсия вычисляются по активациям нейронов, независимо для каждого объекта.

```python
# полезные функции: .std(dim), .mean(dim)
torch.cuda.empty_cache() 
class LayerNormalization(nn.Module):
    def forward(self, x):
        ##### YOUR CODE HERE #####
        mean = x.mean(1, keepdim = True)
        std = x.std(1, keepdim = True)
        return (x - mean)/(std + 1e-8)
        
```

*Выводы:*

Сравнивая графики функций для Batch Normlization и Layer Normalization для Train(первые из двух графиков), можно видеть, что график для BatchNorm более пологий. При BatchNorm статистика вычисляется по батчу и одинакова для каждого примера в батче.Напротив, при LayerNorm статистика вычисляется для каждого объекта и не зависит от других примеров. Независимость между входами означает, что каждый вход имеет различную нормализацию, что позволяет использовать произвольные размеры минибатча. Это объясняет более быструю сходимость при LayerNorm, так при BatchNorm для лучшей нормализации нам необходимо подобрать оптимальный BatchNorm size.
